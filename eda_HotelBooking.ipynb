```python
pd.crosstab(df["hotel"], df['is_canceled'], normalize=True).plot(kind='bar', stacked=True)
```

* **The proportion of cancellations is seen to be more at City Hotel**

```python
sns.boxplot(data=df, x='is_canceled', y='lead_time')
```

* **lead_time is observed to be more for bookings that were cancelled**

```python
ct=pd.crosstab(df['previous_cancellations'],df['is_canceled'], normalize='index')
ct.plot.bar(stacked=True)
```

* **is_canceled is related to previous cancellation especially when previous cancellation is 1 or greater than 13**

* **this may be further studied to understand which distribution channel contributes to the cancellations**

* **if observable segments exist then appropariate measure can be taken**

```python
ct=pd.crosstab([df['previous_cancellations'],df['distribution_channel']],df['is_canceled'], normalize='index')
ct.plot.bar(stacked=True)
```

* **Travel Agents being source of highest bookings also appear to be source of higher cancellations**

```python
plt.figure(figsize=(20,10))
sns.scatterplot(data=df, x='days_in_waiting_list', y='booking_changes', hue='is_canceled' )
#booking changes has also been added to the analysis
```

* **Booking changes is not related to is_canceled**
* **As the number of days_in_waiting_list increases, cancelled observations are found**

```python
plt.figure(figsize=(15,8))
sns.boxplot(y=df_adr_no_outlier['distribution_channel'], x=df_adr_no_outlier['adr'], hue=df['hotel'])
#The variation is split hotelwise also gain more information
```

* **Corporate prices are low as compared to the prices offered to Direct booking and TA/TO bookings**
* **It appears that City hotel charges more than Resort Hotel does**
* **Earlier we observed that City Hotel has a higher proportion of cancellations as opposed to Resort Hotel. Customer prefrence for better price elsewhere, could be a reason**

```python
plt.figure(figsize=(15,8))
sns.boxplot(data=df_adr_no_outlier, x='market_segment', y='adr')
```

* **Online Travel Agents form a big portion of the bookings from the TA/TO distribution channel**


```python
plt.figure(figsize=(15,8))
sns.lineplot(data=df_adr_no_outlier, x='arrival_date_month', y='adr', hue='hotel', sort=False)
#data without the one extreme adr outlier is used here
```

* **Across the months, the price variation at City Hotel is stable than the immense fluctuations at Resort Hotel**

```python
plt.figure(figsize=(15,8))
sns.countplot(data=df_adr_no_outlier, x='arrival_date_month', hue='hotel')
#data without the one extreme adr outlier is used here
```

* **April to August is the peak time with bookings in both the hotels**
* **Resort hotel appears to charge according to season while City hotel maintains a more stable price**

```python
plt.figure(figsize=(15,8))
sns.lineplot(data=df, x='arrival_date_month', y='is_canceled', hue='hotel', sort=False)
```

* **cancellations seem to be higher when bookings are higher, i.e. during the April-August period**

```python
pd.crosstab(df['is_canceled'],df['total_of_special_requests'], normalize='columns')
```

* **If a special request is placed then the probability of cancellation falls**

```python
filter_roomtype = df['reserved_room_type'] != df['assigned_room_type']
df[filter_roomtype]['is_canceled'].value_counts(normalize=True)
```

* **Difference in assigned room type has no effect on cancellation**

```python
pd.crosstab(df['deposit_type'],df['is_canceled'], normalize='index').plot(kind='bar', stacked=True)
```

* **99% of the Non-Refundable depost bookings were cancelled**
